```python
#Rent calculator
#Total room rent, water bill ,Electricity bill, food

rent = int(input("Enter total amount of your flat/room rent : "))
waterBill = int(input("Enter total amount of your water bill : "))
electricity_spend = int(input("Enter total spended electricity : "))
charge_per_unit = int(input("Enter charge per unit on electricity bill : "))
food = int(input("Enter total expenditure on food : "))
persons = int(input("Enter total person in your room : "))

elctricity_bill = electricity_spend * charge_per_unit

#Calculate total bill
Total = (rent + waterBill + food + elctricity_bill ) // persons

print("Total bill have to pay per person :" , Total)
```
